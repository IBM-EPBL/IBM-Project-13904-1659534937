## 5. Create an array of all the even integers from 20 to 35

```python
array_even_20_35 = [x for x in range(20, 35+1) if x % 2 ==0]
print("Array of all the even intergers from 20 to 35 is ", array_even_20_35)
```

## 6. Create a 3x3 matrix with values ranging from 0 to 8

## 7. Concatinate a and b 
## a = np.array([1, 2, 3]), b = np.array([4, 5, 6])

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.concatenate((a, b))
print("Concatenated array is ", c)
```

## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
Dates_1_1_2023_to_10_2_2023 = pd.date_range(start = "1-1-2023", end = '10-2-2023')
print("Series of dates from 1st Jan, 2023 to 10th Feb, 2023 are ", Dates_1_1_2023_to_10_2_2023)
```
